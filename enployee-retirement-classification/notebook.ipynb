```python
import pandas as pd
```

```python
df = pd.read_csv("employee_train.csv")
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
target_col = "leaving"
feature_cols = ["age", "job_satisfaction", "environment_satisfaction", "over_time", "income"]
```

```python
y_train = df[target_col]
X_train = df[feature_cols]

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
df["pred_leaving"] = model.predict(X_train)
df[["leaving", "pred_leaving"]].head()
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(df[target_col], df["pred_leaving"])
print(f"Accuracy: {accuracy}")
```

```python
df_pred = pd.read_csv("employee_pred.csv")
df_pred.head()
```

```python
X_pred = df_pred[feature_cols]
df_pred["pred_leaving"] = model.predict(X_pred)
df_pred.head()
```
